```python
import requests

# Function to get weather data using OpenWeatherMap API
def get_weather(zip_code, api_key):
    # URL to fetch the weather data (using OpenWeatherMap API)
    url = f"http://api.openweathermap.org/data/2.5/weather?zip={zip_code}&appid={api_key}&units=metric"
    
    try:
        # Send GET request to OpenWeatherMap API
        response = requests.get(url)
        data = response.json()
        
        # Check if the response contains valid weather data
        if response.status_code == 200:
            main = data['main']
            weather = data['weather'][0]
            temperature = main['temp']
            humidity = main['humidity']
            description = weather['description']
            
            # Print the weather details
            print(f"Weather for {zip_code}:")
            print(f"Temperature: {temperature}°C")
            print(f"Humidity: {humidity}%")
            print(f"Condition: {description.capitalize()}")
        else:
            print(f"Error: {data['message']}")
    except requests.exceptions.RequestException as e:
        print(f"Error fetching weather data: {e}")

# Function to fetch current weather based on zip code
def fetch_weather():
    # Get the API key (replace 'your_api_key' with your actual OpenWeatherMap API key)
    api_key = "your_api_key"   # need to add your api key here to be able to see the weather forecast 
    
    # Ask the user for zip/postal code
    zip_code = input("Enter zip/postal code: ")

    # Fetch the weather for the given zip code
    get_weather(zip_code, api_key)

# Function to fetch weather for user's location (optional)
def fetch_weather_for_location():
    try:
        # Get the user's current IP address location
        ip_url = "https://api.ipify.org?format=json"
        ip_response = requests.get(ip_url)
        ip_data = ip_response.json()
        ip_address = ip_data["ip"]
        
        # Use the IP to fetch geolocation info
        location_url = f"https://ipinfo.io/{ip_address}/json"
        location_response = requests.get(location_url)
        location_data = location_response.json()
        city = location_data["city"]
        country = location_data["country"]
        print(f"Fetching weather for your location: {city}, {country}")

        # Get weather for the city
        weather_url = f"http://api.openweathermap.org/data/2.5/weather?q={city},{country}&appid={api_key}&units=metric"
        weather_response = requests.get(weather_url)
        weather_data = weather_response.json()
        
        if weather_response.status_code == 200:
            main = weather_data['main']
            weather = weather_data['weather'][0]
            temperature = main['temp']
            humidity = main['humidity']
            description = weather['description']
            
            # Print the weather details
            print(f"Weather for {city}, {country}:")
            print(f"Temperature: {temperature}°C")
            print(f"Humidity: {humidity}%")
            print(f"Condition: {description.capitalize()}")
        else:
            print(f"Error: {weather_data['message']}")
    except requests.exceptions.RequestException as e:
        print(f"Error fetching location or weather data: {e}")

# Main function to fetch weather data
def main():
    choice = input("Would you like to enter a zip/postal code (1) or fetch weather for your location (2)? Enter 1 or 2: ")

    if choice == '1':
        fetch_weather()
    elif choice == '2':
        fetch_weather_for_location()
    else:
        print("Invalid choice. Please enter 1 or 2.")

if __name__ == "__main__":
    main()
```
