```python
#we will create graphs from links here, it will be a directed graph
import matplotlib.pyplot as plt #importing matplotlib library
import networkx as nx   #importing networkx library

#creating a graph
def create_graph(links):
    G = nx.DiGraph() #creating a directed graph
    for link in links:
        node1,node2 = link
        G.add_edge(node1,node2) #adding edges to the graph
    return G

#plotting the graph
def visualize_graph(G):
    plt.figure(figsize=(10,10)) #setting the size of the figure
    nx.draw(G,with_labels=True, node_color= 'skyblue', node_size   = 2000, font_size= 12, font_weight = 'bold', edge_color ='gray')
    plt.title("Graph of the links") #setting the title of the graph
    plt.show() #displaying the graph
    
    
    #example list fo links 
links = [('utku', 'miyeon'),('miyeon','noa'),('noa','utku'),('noa','2'),('2','C'),('E','C'),('E','A'),('A','D'),('D','B'),('D','E'),('C','D'),('F','B'),('B','F')]

G = create_graph(links)

visualize_graph(G)
```
